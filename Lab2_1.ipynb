# **1.10 สร้าง data batch โดยมีความกว้าง 64 samples**
- print แสดงข้อมูลทีละ batch โดยมี ข้อความหัวข้อแสดงให้รู้ทุก batch


```python
nl=(len(train_x)/64)+1
block=[]
batch=[]
for i in range(int(nl)):
  block = train_x[64*i:64*(i+1)]
  batch.append(block)
  print("Batch : ",i+1)
  print("Shape : ",batch[i].shape)
```

แสดงตัวอย่างการดูข้อมูลที่ batch[5] และ batch[6] ที่บรรจุข้อมูลแต่ละช่วงไว้

```python
print(batch[5])
print("-----------------------------\n",len(batch[5]))
```

```python
print(batch[6])
print("-----------------------------\n",len(batch[6]))
```

**- ยืนยันผลด้วยว่าทุกรอบมี 64 samples จริง (ทำยังไง คิดเอง)**

ตอบ จะเห็นได้ว่า สามารถนำ batch[0-6]  มารันเพื่อแสดงข้อมูล และสามารถตรวจสอบขนาดได้ทั้ง len และ shape ของแต่ละ batch จะได้จำนวนข้อมูลเท่ากับ 64 บวกกับ batch สุดท้ายที่จัดเก็บ 20 ตัว
