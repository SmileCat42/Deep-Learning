เขียน Naive implementation และแสดงผลการเทียบค่า naive_dot(a, b) กับ np.dot(a, b)


```python
def naive_vector_dot(x, y):
  assert len(x.shape) == 1
  assert len(y.shape) == 1
  assert x.shape[0] == y.shape[0]

  z = 0.
  for i in range(x.shape[0]):
    z += x[i] * y[i]
  return z

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(naive_vector_dot(a, b))
np.dot(a, b)
```

12.	หากต้องการให้ผลลัพธ์มีรูปร่าง (3×2) ต้องออกแบบขนาดของ B ใหม่อย่างไร

ตอบ  ถ้าอยากเปลี่ยนให้ ผลลัพธ์มีขนาด (3x2) ก็สามารถเปลี่ยนมิติที่ 2 ของ B ให้เหลือ 2 ได้โดยการลบ 3 และ 6 ออก จะได้ [1,2],[4,5] ที่มีขนาด (2x2) แต่สำหรับมิติแรกของ B ไม่สามารถเปลี่ยนได้ เนื่องจากจำเป็นต้องทำให้ขนาดเท่ากับมิติที่ 2 ของ A ซึ่งก็คือ 2 ถึงจะสามารถนำมา dot กันได้

```python
x = np.array([[0.,1.],
[2.,3.],
[4.,5.]])
x1 = x.reshape((6,1))
x2 = x.reshape((2,3))

print(x1)
print(x2)
```
